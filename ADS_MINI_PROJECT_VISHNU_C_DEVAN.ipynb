### Data Wrangling

```python
#Merging:
# Combine two datasets (e.g., a subset or external data) using keys.
# Example: Merge the dataset with itself to demonstrate functionality
df_subset = df[['EmployeeNumber', 'Age', 'MonthlyIncome']]
merged_data = pd.merge(df, df_subset, on='EmployeeNumber', suffixes=('', '_copy'))
print(merged_data.head())
```

```python
#Concatenation:
#Append datasets either row-wise or column-wise.

# Concatenate subsets of columns
data_part1 = df[['Age', 'MonthlyIncome']]
data_part2 = df[['Attrition', 'DistanceFromHome']]
concatenated_data = pd.concat([data_part1, data_part2], axis=1)
print(concatenated_data.head())
```

```python
#Pivoting:
#Summarize data into a table format.
# Pivot table to show average Monthly Income by Department and Attrition
pivot_table = df.pivot_table(values='MonthlyIncome', index='Department', columns='Attrition', aggfunc='mean')
print(pivot_table)
```

```python
#Reshaping:
#Transform data from wide to long format.
# Reshape data into long format
reshaped_data = pd.melt(df, id_vars=['EmployeeNumber', 'Attrition'], value_vars=['Age', 'MonthlyIncome'], var_name='Metric', value_name='Value')
print(reshaped_data.head())
```

```python
# Grouping and Aggregation
#Aggregate statistics based on groupings
#Group by Attrition and calculate average age and income
grouped_data = df.groupby('Attrition').agg({'Age': 'mean', 'MonthlyIncome': 'mean'})
print(grouped_data)
```

```python
#Crosstab:
#Create a cross-tabulation for categorical variables.
# Crosstab of Attrition and Department
crosstab_result = pd.crosstab(df['Attrition'], df['Department'])
print(crosstab_result)
```

```python
#Data Transformations
#Binning:
#Divide continuous data into intervals.
# Binning Age into categories
bins = [18, 30, 40, 50, 60]
labels = ['18-30', '31-40', '41-50', '51-60']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
print(df[['Age', 'AgeGroup']].head())
```

```python
#Encoding:
#Convert categorical data into numeric format.
# Encoding Attrition as binary
df['AttritionEncoded'] = df['Attrition'].map({'Yes': 1, 'No': 0})
print(df[['Attrition', 'AttritionEncoded']].head())
```

```python
#Standardization:
#Scale numerical columns to have a mean of 0 and standard deviation of 1.
from sklearn.preprocessing import StandardScaler

# Standardize numerical columns
scaler = StandardScaler()
numerical_cols = ['Age', 'MonthlyIncome', 'DistanceFromHome']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
print(df[numerical_cols].head())
```

### Final Visualization and Insights

```python
#Advanced Visualizations
#Subplots: Attrition Analysis Across Multiple Metrics
#This visualization compares attrition rates across various factors such as age groups, income levels, and job roles.
```

```python

# Subplots for key metrics
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Attrition by Age Group
sns.countplot(x='AgeGroup', hue='Attrition', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Attrition by Age Group')
axes[0, 0].set_xlabel('Age Group')
axes[0, 0].set_ylabel('Count')

# Monthly Income Distribution by Attrition
sns.kdeplot(data=df, x='MonthlyIncome', hue='Attrition', fill=True, ax=axes[0, 1])
axes[0, 1].set_title('Monthly Income Distribution by Attrition')
axes[0, 1].set_xlabel('Monthly Income')
axes[0, 1].set_ylabel('Density')

# Attrition by Job Role
sns.countplot(y='JobRole', hue='Attrition', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Attrition by Job Role')
axes[1, 0].set_xlabel('Count')
axes[1, 0].set_ylabel('Job Role')

# Distance From Home Distribution by Attrition
sns.boxplot(x='Attrition', y='DistanceFromHome', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Distance From Home by Attrition')
axes[1, 1].set_xlabel('Attrition')
axes[1, 1].set_ylabel('Distance From Home')

plt.tight_layout()
plt.show()
```

```python
#Annotations and Insights
#Correlation Heatmap with Annotations
#A heatmap showcasing correlations between features and attrition.
# Correlation heatmap
numeric_data = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True, vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

Interpretations and Actionable Insights

Subplot Insights:

Attrition by Age Group:

Younger employees (18-30 age group) show higher attrition rates compared to older employees.

Insight: Implement targeted retention programs for younger employees, such as career development opportunities.

Monthly Income Distribution:

Attrition is more frequent among employees with lower monthly income.
Insight: Consider revising compensation structures, especially for employees in lower income brackets.

Attrition by Job Role:

Specific job roles (e.g., Sales Representatives) have notably higher attrition rates.

Insight: Investigate job satisfaction, workload, and support for high-attrition roles.

Distance From Home:

Employees living farther from work tend to leave more frequently.

Insight: Explore flexible work arrangements or relocation assistance for distant employees.

Correlation Heatmap:

Positive Correlation: Monthly Income and Total Working Years are positively correlated.

Negative Correlation: Attrition has a mild negative correlation with Job Satisfaction and Environment Satisfaction.

Insight: Enhancing workplace satisfaction metrics can potentially reduce attrition rates.
